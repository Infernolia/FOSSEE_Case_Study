```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv("TREE.csv")
```

```python
df
```

```python
for i in df.columns[1:17]:
    
    x = df[i].corr(df['SO2'])
    y = df[i].corr(df['NOx'])
    z = df[i].corr(df['AQI'])
    a = df[i].corr(df['PM10'])
    b = df[i].corr(df['Other_Tree_Count'])
    print ("\n\n" + str(i) + "\n SO2: " + str(x) + " \n NOx: " + str(y) + "\n AQI: " + str(z) + "\n  PM10: " + str(a) + "\n Other Tree Count: " + str(b))
```
