```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('doctor_train.csv')
test = pd.read_csv('doctor_test.csv')
```

```python
test.info()
```

```python
train = train.dropna()
```

```python
test.drop(columns=["communication","Irregular","age","Money"],inplace=True)

train.drop(columns=["communication","Irregular","age","Money"],inplace=True)
train.set_index(train['ID'])
train.drop(columns=["ID"],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
train['Profession']= le.fit_transform(train['Profession']) 
train['Status']= le.fit_transform(train['Status']) 
train['Month']= le.fit_transform(train['Month'])



train['side_effects']= le.fit_transform(train['side_effects']) 
train['Y']= le.fit_transform(train['Y']) 

  
test['Profession']= le.fit_transform(test['Profession']) 
test['Status']= le.fit_transform(test['Status']) 
test['Month']= le.fit_transform(test['Month'])



test['side_effects']= le.fit_transform(test['side_effects']) 
```

```python
for x in train.index:
    if(train.at[x,"edu"]=="primary"):
        train.at[x,"edu"]=0.0
    elif (train.at[x,"edu"]=="secondary"):
        train.at[x,"edu"]=1.0
    elif (train.at[x,"edu"]=="tertiary"):
        train.at[x,"edu"]=2.0
    else:
        train.at[x,"edu"]=3.0
        
        
for x in test.index:
    if(test.at[x,"edu"]=="primary"):
        test.at[x,"edu"]=0.0
    elif (test.at[x,"edu"]=="secondary"):
        test.at[x,"edu"]=1.0
    elif (test.at[x,"edu"]=="tertiary"):
        test.at[x,"edu"]=2.0
    else:
        test.at[x,"edu"]=3.0
```

```python
for x in train.index:
    if(train.at[x,"residence"]=="yes"):
        train.at[x,"residence"]=0.0
    elif (train.at[x,"residence"]=="no"):
        train.at[x,"residence"]=1.0
    else:
        train.at[x,"residence"]=2.0
        
        
for x in train.index:
    if(train.at[x,"prev_diagnosed"]=="yes"):
        train.at[x,"prev_diagnosed"]=0.0
    elif (train.at[x,"prev_diagnosed"]=="no"):
        train.at[x,"prev_diagnosed"]=1.0
    else:
        train.at[x,"prev_diagnosed"]=2.0
        
        
        
for x in test.index:
    if(test.at[x,"residence"]=="yes"):
        test.at[x,"residence"]=0.0
    elif (test.at[x,"residence"]=="no"):
        test.at[x,"residence"]=1.0
    else:
        test.at[x,"residence"]=2.0
        
        
for x in test.index:
    if(test.at[x,"prev_diagnosed"]=="yes"):
        test.at[x,"prev_diagnosed"]=0.0
    elif (test.at[x,"prev_diagnosed"]=="no"):
        test.at[x,"prev_diagnosed"]=1.0
    else:
        test.at[x,"prev_diagnosed"]=2.0
```
