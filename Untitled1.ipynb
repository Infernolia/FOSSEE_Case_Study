```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np

df = pd.read_csv('India_indicators.csv')
%matplotlib notebook
```

```python
df
```

```python
df = df.rename(columns ={'Unnamed: 0': 'Year'})
```

```python
df
```

```python
l = df.columns[1:4]
X_clustering = df[l]
Nc = range(1, 50)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(X_clustering).score(X_clustering) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
model = KMeans(n_clusters = 5 )
model.fit(X_clustering)
print(model.labels_)
colormap = np.array(['Red', 'Blue', 'Green', 'Orange', 'Pink'])
plt.scatter(X_clustering['GDP (current US$)'], X_clustering['Death rate, crude (per 1,000 people)'], c = colormap[model.labels_])
```

```python
l = model.labels_
```

```python
L = []
for i in range(0,50):
    print("\n Year" + str(df.at[i,'Year']) + " " + str(l[i]))
    if df.at[i,'Year']== 1974 or df.at[i,'Year']== 1994 or df.at[i,'Year']== 2002 or df.at[i,'Year']== 2003 or df.at[i,'Year']== 2004 or df.at[i,'Year']== 2006 or df.at[i,'Year']== 2014 or df.at[i,'Year']== 2015 or df.at[i,'Year']== 2018:
        L.append(1)
    else:
        L.append(0)
```

```python


#create new df 
df1 = pd.DataFrame({'Label':L})
df1
```

```python
x = pd.concat([df, df1.reindex(df.index)], axis=1)
```

```python
x
```

```python
X = x.drop(columns = 'Label')
y = x['Label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
accurarices = cross_val_score(classifier, X_train, y_train, cv= 5)
accurarices
```
