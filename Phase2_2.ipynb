```
import pandas as pd
import numpy as np
audio = pd.read_csv("/content/drive/My Drive/Awaaz/audio.csv",encoding='latin1')
episodes = pd.read_csv("/content/drive/My Drive/Awaaz/episodes.csv",encoding='latin1')
podcast = pd.read_csv("/content/drive/My Drive/Awaaz/podcast.csv",encoding='latin1')
```

```
episodes = episodes[['show_id','episode_id']]
```

```
e = episodes['show_id'].value_counts().reset_index()
```

```
e
```

```
e.columns = ['show_id','no_of_eps']
```

```
e
```

```
audio = audio[['aawaz_user_id','episode_id','show_id']]
```

```
audio
```

```
a1 = audio.groupby(['aawaz_user_id','show_id']).count()
```

```
a1
```

```
a1.to_csv('trial.csv')
a1 = pd.read_csv('trial.csv')
a1
```

```
a1['completion'] = 0
```

```
for i in a1.index:
  for j in e.index:
    if a1.at[i,"show_id"]==e.at[j,"show_id"]:
      a1.at[i,'completion'] = a1.at[i,'episode_id']/e.at[j,'no_of_eps']
      break
```

```
a1
```

```
a1.drop(columns=['episode_id'],inplace=True)
a1
```

```
x = a1.groupby(['aawaz_user_id','show_id'])['completion'].mean()
x.columns = ['aawaz_user_id','show_id','avg_comp_rate']
x
```

```
x.to_csv("a.csv")
x = pd.read_csv("a.csv")
x.columns = ['aawaz_user_id','show_id','avg_comp_rate']
x
```
