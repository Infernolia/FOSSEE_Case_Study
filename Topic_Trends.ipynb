```
df = pd.read_csv('/content/drive/My Drive/event.csv')
```

```
df.drop(columns = ["Unnamed: 0"],inplace = True)
df
```

```
for i in df.index:
  if df.at[i,"month"]<10:
    df.at[i,"Date"] = str(df.at[i,"year"]) + "0" + str(df.at[i,"month"])
  else:
    df.at[i,"Date"] = str(df.at[i,"year"]) + str(df.at[i,"month"])
  df.at[i,"Date"] = int(df.at[i,"Date"])
df
```

```
sorted = df.sort_values('Date',ascending = True)
sorted
```

```
topic = "coronavirus"
sorted["count"] = 0

for i in sorted.index:
  if sorted.at[i,"Topic"].count(topic)>0:
    sorted.at[i,"count"] = sorted.at[i,"count"] + 1
```

```
import nltk
from textblob import TextBlob
```

```
sorted["Subjectivity"] = 0.0
sorted["Polarity"] = 0.0

for i in sorted.index:
  text = str(sorted.at[i,"without_emoji"])
  obj = TextBlob(text)
  sorted.at[i,"Subjectivity"] = float(obj.subjectivity)
  sorted.at[i,"Polarity"] = float(obj.polarity)

sorted
```

```
from pandas import DataFrame

topic = 'people'
dates = []
subj = []
polarity = []

for i in sorted.index:
  if sorted.at[i,"Topic"].count(topic)>0:
    dates.append(sorted.at[i,"Date"])
    subj.append(sorted.at[i,"Subjectivity"])
    polarity.append(sorted.at[i,"Polarity"])

arr = []

arr.append(dates)
arr.append(subj)
arr.append(polarity)
df = DataFrame (arr).transpose()
df.columns = ['Date','Subjectivity','Polarity']

df = df.melt('Date', var_name='cols',  value_name='vals')
g = sns.factorplot(x="Date", y="vals", hue='cols', data=df)


```

```
from matplotlib import pyplot

topic = 'quarantine'

sorted["count"] = 0

for i in sorted.index:
  if sorted.at[i,"Topic"].count(topic)>0:
    sorted.at[i,"count"] = sorted.at[i,"count"] + 1

a4_dims = (20, 6)
fig, ax = pyplot.subplots(figsize=a4_dims)
ax.set_title(topic)
sns.lineplot(ax = ax, x="Date", y="count", data=sorted)


dates = []
subj = []
polarity = []

for i in sorted.index:
  if sorted.at[i,"Topic"].count(topic)>0:
    dates.append(sorted.at[i,"Date"])
    subj.append(sorted.at[i,"Subjectivity"])
    polarity.append(sorted.at[i,"Polarity"])

arr = []

arr.append(dates)
arr.append(subj)
arr.append(polarity)
df = DataFrame (arr).transpose()
df.columns = ['Date','Subjectivity','Polarity']

try:
  a4_dims = (20, 6)
  fig, ax = pyplot.subplots(figsize=a4_dims)
  ax.set_title(topic)
  df = df.melt('Date', var_name='cols',  value_name='vals')
  sns.lineplot(ax =ax, x="Date", y="vals", hue='cols', data=df)

except:

  print("empty dataframes")
```
