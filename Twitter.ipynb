```
celebdf['month'] = 0
celebdf['year'] = 0
for i in celebdf.index:

  datestring = str(celebdf.at[i,'date'])
  try:
    dt = datetime.strptime(datestring, '%Y-%m-%d')
  except:
    dt = datetime.strptime(datestring, '%d-%m-%Y')
  celebdf.at[i,'year'] = dt.year
  celebdf.at[i,'month'] = dt.month

celebdf
```

```
keydf['month'] = 0
keydf['year'] = 0
for i in celebdf.index:
  keydf.at[i,'date'] = str(keydf.at[i,'date']).split(" ")[0]
  

keydf
```

```

keydf['month'] = 0
keydf['year'] = 0

for i in keydf.index:
  
  if str(keydf.at[i,'date']).count(':')>=1:
     keydf.at[i,'date'] = str(keydf.at[i,'date']).split(" ")[0]
     


keydf   
```

```
keydf['month'] = 0
keydf['year'] = 0

keydf.dropna(subset = ["date"], inplace=True)
for i in keydf.index:

  datestring = str(keydf.at[i,'date'])
  try:
    dt = datetime.strptime(datestring, '%Y-%m-%d')
  except:
    dt = datetime.strptime(datestring, '%d-%m-%Y')
  keydf.at[i,'year'] = dt.year
  keydf.at[i,'month'] = dt.month

keydf
```

```
eventdf['month'] = 0
eventdf['year'] = 0

eventdf.dropna(subset = ["date"], inplace=True)
for i in eventdf.index:

  datestring = str(eventdf.at[i,'date'])

  if datestring.count(" ") >=1 :
    datestring = datestring.split(" ")[0]
  try:
    try:
      dt = datetime.strptime(datestring, '%Y-%m-%d')
      eventdf.at[i,'year'] = dt.year
      eventdf.at[i,'month'] = dt.month
    except:
      dt = datetime.strptime(datestring, '%d-%m-%Y')
      eventdf.at[i,'year'] = dt.year
      eventdf.at[i,'month'] = dt.month
      
  except:
    x = 2020
    y = 6
    eventdf.at[i,'year'] = x
    eventdf.at[i,'month'] = y

  

eventdf
```
