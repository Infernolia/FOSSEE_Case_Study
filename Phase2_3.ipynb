```
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import numpy as np
audio = pd.read_csv("/content/drive/My Drive/Awaaz/phase2_1.csv",encoding='latin1')
episodes = pd.read_csv("/content/drive/My Drive/Awaaz/episodes.csv",encoding='latin1')
podcast = pd.read_csv("/content/drive/My Drive/Awaaz/podcast.csv",encoding='latin1')
```

```
audio
```

```
audio.drop(columns=['Unnamed: 0'],inplace=True)
audio
```

```
audio = audio.rename(columns ={'Unnamed: 12': 'anon'})
audio
```

```
names = []
for i in  audio['subcategory'].unique():
  if pd.isnull(i):
    names.append('anon')
  else:
    names.append(i)
```

```
names
```

```
groups = [[]]

for i in  names:
  l = []
  try:
    temp = audio[audio[i]==1]
    for j in temp.index:
      l.append(temp.at[j,"aawaz_user_id"])
  except:
    print(i)
  groups.append(l)
```

```
groups
```

```
df = pd.DataFrame(groups)
df = df.transpose()
```

```
df
```

```
df = df.drop(columns = [0])
```

```
df.columns = names
```
