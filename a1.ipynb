```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
data = pd.read_csv("next.csv")
```

```python
data
```

```python
data['Clade'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red'
)
```

```python
col_one_list = data['Country'].unique().tolist()
```

```python
col_one_list
```

```python
data_cfr = pd.read_csv("coronavirus-cfr.csv")
```

```python
fin_cfr= data_cfr[data_cfr['Entity'].isin(col_one_list)]
```

```python
fin_cfr
```

```python
#case fatality rate
fin_cfr['Entity'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red'
)
```

```python
data_beds = pd.read_csv("hospital-beds-per-1000-people.csv")
fin_beds= data_beds[data_beds['Entity'].isin(col_one_list)]
fin_beds['Entity'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red'
)
```

```python

sns.barplot(x=fin_beds["Hospital beds (per 100,000)"], y=fin_beds["Entity"])
```

```python
data_doctors = pd.read_csv("physicians-per-1000-people.csv")
fin_doctors= data_doctors[data_doctors['Entity'].isin(col_one_list)]
fin_doctors['Entity'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red'
)
```

```python

sns.barplot(x=fin_doctors["doc"], y=fin_doctors["Entity"])
```

```python
data_age = pd.read_csv("share-of-the-population-that-is-70-years-and-older.csv")
fin_age= data_age[data_age['Entity'].isin(col_one_list)]
fin_age['Entity'].value_counts().sort_index().plot.bar(
    figsize=(12, 6),
    color='red'
)
```

```python

sns.barplot(x=fin_age["Share who is 70 or over (%)"], y=fin_age["Entity"])
```
